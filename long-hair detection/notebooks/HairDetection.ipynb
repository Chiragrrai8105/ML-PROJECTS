```python
import os

dataset_path = r"D:\dataset\p1\Hair"

train_images = []
valid_images = []

# Load training images
train_path = os.path.join(dataset_path, "train", "images")

for file in os.listdir(train_path):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        train_images.append(os.path.join(train_path, file))

# Load validation images
valid_path = os.path.join(dataset_path, "valid", "images")

for file in os.listdir(valid_path):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        valid_images.append(os.path.join(valid_path, file))

print("Training Images:", len(train_images))
print("Validation Images:", len(valid_images))
print("\nSample Training Image:")
print(train_images[0])
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data=r"D:\dataset\p1\Hair\data.yaml",
    epochs=100,
    imgsz=640,
    batch=16
)
```

```python
metrics = model.val()
```
